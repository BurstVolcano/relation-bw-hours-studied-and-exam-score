```python
import pandas
import matplotlib.pyplot as plt

data = pandas.read_csv('StudentPerformanceFactors.csv')

plt.scatter(data['Hours_Studied'], data['Exam_Score'])
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Hours Studied vs Exam Score')
plt.show()

#This correlation is used to identify if there is any relation between hours and score.
#If correlation is close to 0 then there is no relation between them and if it is close to 1 they have strong relationship.
correlation = data['Hours_Studied'].corr(data['Exam_Score'])
print(f"Correlation between Hours Studied and Exam Score: {correlation}")
```

**From the correlation it is clear that the value is moderate.**

So,  A value of 0.45 means that as study hours increase, exam scores tend to increase as well. However, this relationship is not very strong, meaning that while study hours do have an impact, they aren’t the only factor affecting exam scores.
